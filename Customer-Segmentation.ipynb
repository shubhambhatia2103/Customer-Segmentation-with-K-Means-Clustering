#### **พื้นฐาน Clustering (Unsupervised Learning)**
การแบ่งกลุ่ม (clustering) เป็นการวิเคราะห์ทาง data science เพื่อจำแนกข้อมูลออกเป็นกลุ่ม ๆ และนำกลุ่มที่แบ่งได้ไปใช้ในงาน เช่น การแบ่งกลุ่มลูกค้าที่มาซื้อสินค้า เมื่อ Data Scientist ได้จำแนกกลุ่มลูกค้า ทางฝ่ายการตลาด สามารถออกแผนการตลาด ให้ตรงกับกลุ่มเป้าหมาย เป็นการช่วยลดค่าใช้จ่าย และ ไม่เป็นการรบกวนลูกค้า ที่ไม่ใช่กลุ่มเป้าหมายด้วย

การแบ่งกลุ่มออกเป็น cluster มีหลักง่าย ๆ คือ ข้อมูลที่มีความคล้ายกัน ถูกจัดอยู่ใน cluster เดียวกัน เช่น ลูกค้าที่มาซื้อสินค้าในซุปเปอร์ เรากำหนดให้ แบ่งลูกค้าออกเป็น 2 กลุ่ม วิธีง่ายที่สุด คือ แบ่งตามเพศ ทำให้เราได้ กลุ่มผู้ซื้อเพศชาย กลุ่มผู้ซื้อเพศหญิง เป็นต้น

  - Hard Clustering / การกำหนดให้แต่ละชุดข้อมูล แบ่งออกเป็นกลุ่มที่แยกออกจากกันโดยสิ้นเชิง
  - Soft Clustering / ข้อมูลมีโอกาสที่จะอยู่ในหลายๆ กลุ่มได้ ขึ้นอยู่กับความน่าจะเป็นของตัวข้อมูล

### **2. ประเภทของ Clustering**

#### **Partitioning Clustering**
เทคนิค clustering แรก คือ การแบ่งเป็นส่วน ๆ (partitioning) โดย data point จะถูกแบ่งออกเป็น $k$ กลุ่ม ซึ่ง Data Scientist เป็นผู้กำหนดขึ้นมา (อาจใช้การ plot กราฟเพื่อหา elbow หรือ จากการวิเคราะห์ข้อมูล เราจะอธิบายเพิ่มเติมในภายหลัง)

สำหรับเทคนิค partitioning ที่นิยม คือ $K$-means โดยแต่ละ cluster จะมีศูนย์กลางหรือค่าเฉลี่ย เป็นตัวแทนของกลุ่ม $K$-means เป็นอัลกอริทึมที่ ลดผลรวมข้อผิดพลาดยกกำลังสอง (total squared error) จากตำแหน่งศูนย์กลางของแต่ละ cluster

ข้อดีของ $K$-means คือ ใช้งานง่าย ไม่ยุ่งยากซับซ้อน แต่ข้อเสีย คือ มีความไวต่อข้อมูลที่ผิดปกติ และ ข้อมูลที่เป็น outlier ดังนั้น ประสิทธิภาพของเทคนิคนี้ จึงขึ้นอยู่กับ criteria ที่ใช้ในการแบ่งกลุ่ม

อีกวิธีที่มีใน partitioning คือ $K$-medoids ซึ่งมีข้อดี คือ เมื่อเทียบกับ $K$-means เทคนิคนี้มีความไวต่อ outlier น้อยกว่า อัลกอริทีมที่ใช้ คือ ลดผลรวมของความแตกต่าง ระหว่าง data point ที่มีป้ายกำกับให้อยู่ใน cluster และ data point ที่ถูกกำหนดให้เป็นตัวแทน cluster

#### **Hierarchy Clustering**
การจัดกลุ่มตามลำดับชั้น (hierarchy clustering) เป็นการแบ่งกลุ่ม โดยไม่ต้องมีการกำหนดจำนวนกลุ่ม โดย Data Scientist ตัวอัลกอริทึมจะทำการแบ่ง cluster เอง ซึ่งเป็นการแบ่งที่เรียกว่า โครงต้นไม้ (tree-based cluster) หรือ dendrogram โดยแต่ละ node เป็นตัวแทนของหนึ่ง cluster เราสามารถอธิบายหลักการของ Dendrogram ง่าย ๆ ดังนี้
1. อัลกอริทึมจะหา dataset สองอันที่อยู่ใกล้กันที่สุด
2. ทำการรวม (merge) เข้าด้วยกัน เกิดเป็นกลุ่มขึ้นมา
3. วนกลับไปทำข้อ 1 ใหม่ จนกว่าจะ dataset ทั้งหมดจะถูก merge เข้าด้วยกัน

ข้อดีของ hierarchy clustering คือ สามารถแสดงผลที่เข้าใจง่าย แต่ข้อเสีย คือ ไม่เหมาะสมที่จะใช้กับ dataset ที่มีจำนวนมาก

#### **Fuzzy Clustering**
สำหรับ fuzzy clustering หรือ soft clustering ข้อมูลหนึ่งจุด สามารถถูกจัด ให้เป็นสมาชิกในมากกว่าหนึ่ง cluster จึงเป็นที่มาของคำว่า soft clustering แอพพลิเคชั่นที่นิยมสำหรับ fuzzy clustering คือ งานจำพวก pattern recognition ยกตัวอย่างเช่น การ clustring เอกสาร เนื่องจากเอกสารบางฉบับ เราไม่สามารถเจาะจงได้ว่าให้อยู่ใน cluster ใด cluster หนึ่ง เพราะมีเนื้อหา ที่ทับซ้อนอยู่ระหว่าง cluster เป็นต้น

เทคนิคของ fuzzy ที่นิยมใช้กัน คือ Fuzzy C-means (คุ้น ๆ หรือไม่ครับ "means" ถูกแล้วครับ $K$-means นั่นเอง) สำหรับ Fuzzy C-means นั้น เราต้องกำหนดจำนวน cluster เช่นเดียวกันกับ $K$-means

เราจะทราบได้อย่างไรว่า ต้องแบ่งกี่กลุ่ม จริง ๆ แล้วไม่มีหลักการตายตัว ผู้สอนขอแนะนำว่า ให้ผู้เรียนทำการวิเคราะห์ข้อมูล และลองจินตนาการ ดูว่าควรจะแบ่งประมาณกี่กลุ่ม จากนั้นก็ทดลองโมเดล พอได้ผลแล้วก็ลองดูกลุ่มที่ถูกแบ่งว่าเหมาะสมหรือไม่ ผู้สอนพอมีหลักคร่าว ๆ คือ ถ้าจำนวนข้อมูลไม่มาก เราควรกำหนดจำนวนกลุ่ม ไม่เยอะมาก

Fuzzy C-means มีความคล้ายกันกับ $K$-means โดยมีขั้นตอนการแบ่งกลุ่มดังนี้

* Step 1: กำหนดจำนวน cluster (แนะนำว่าให้เร่ิมจากจำนวนน้อยก่อน)
* Step 2: กำหนดค่า coefficient หรือ degree of membership แบบสุ่มสำหรับ data point ใน cluster นั้น โดยค่า coefficient อยู่ในช่วง 0 ถึง 1 ถ้าค่า coefficient เข้าใกล้ 0 หมายความว่า data point นั้นอยู่ห่างจากจุดศูนย์กลางของ cluster มาก
* Step 3: คำนวนจุดศูนย์กลางของ cluster
* Step 4: คำนวนค่า coefficient ของแต่ละ data point
* Step 5: ทำ step 3 และ 4 จนกว่าจะครบจำนวนลูปที่กำหนดไว้ หรือจนกว่าค่า coefficient มีการเปลี่ยนแปลงต่ำกว่าค่า $\epsilon$ ที่กำหนดไว้

ข้อดีของ Fuzzy C-means คือ เหมาะสำหรับ data set ที่มีการ overlap ของ cluster และ data point สามารถถูกจัดกลุ่มได้มากกว่า 1 cluster

ข้อเสีย คือ ต้องมีการทดลอง เลือกจำนวน cluster ให้เหมาะสม ใช้เวลาในการประมวลผลสูง โดยเฉพาะถ้าค่า $\epsilon$ ยิ่งน้อย ก็ต้องยิ่งวนลูปนานขึ้น


#### **Density-Based Clustering (DBSCAN)**
สำหรับกลุ่มนี้ เป็นการแบ่งกลุ่มข้อมูล โดยการจัดกลุ่มตามความหนาแน่น (density-based) ซึ่งการที่จะเป็น cluster หรือไม่นั้น ขึ้นอยู่กับความหนาแน่นของ data point ในบริเวณรอบ ๆ ในระยะไม่เกินค่าที่กำหนดไว้ ถ้า data point ไหนอยู่นอกขอบเขต ก็จะถือว่าเป็น noise หรือ outlier และไม่ถูกจัดอยู่ในกลุ่มใด

สำหรับอัลกอริทึมที่นิยมใช้ ได้แก่ Density-based spatial clustering of applications with noise (DBSCAN) ซึ่งอัลกอริทึมนี้ จะเชื่อม data point ที่อยู่ใกล้ ๆ กัน ตามเงื่อนไขที่กำหนด

```
# เนื่องจากการทำ clustering คือการหาความคล้ายกัน โดยการวัดระยะห่างของ Data point แตะละจุุด
# เรามาลองใช้สายตาดูว่า tao hong korn ถ้าดูจากข้อมูลที่มี ใครมีความคล้ายกันมากที่สุด

Tao = [26, 30000]
Hong = [40, 34000]
Korn = [25, 35000]
```

```
import pandas as pd
df = pd.DataFrame(data=[Tao, Hong, Korn], columns=['Age', 'Income'])
df
```

```
# สร้าง function ให้เพื่อวัดระยะห่างของ vector 2 ชุด

def distance(vecOne, vecTwo):
    dist = ((vecOne[0]-vecTwo[0])**2 + (vecOne[1] - vecTwo[1])**2)**(1/2)
    return dist
```

```
dis = distance(Tao, Hong)
dis
```

```
dis = distance(Hong, Korn)
dis
```

```
dis = distance(Tao, Korn)
dis
```

```
df_norm = (df - df.mean())/df.std()
df_norm
```

```
# หาความห่างของ tao และ hong ที่มีการ norm feature

nor_dis = distance(df_norm.iloc[0], df_norm.iloc[1])
nor_dis
```

```
# หาความห่างของ tao และ korn ที่มีการ norm feature

nor_dis = distance(df_norm.iloc[0], df_norm.iloc[2])
nor_dis
```

```
# หาความห่างของ hong และ korn ที่มีการ norm feature

nor_dis = distance(df_norm.iloc[1], df_norm.iloc[2])
nor_dis
```

1. กำหนดหรือสุ่มค่าเริ่มต้น จำนวน K ค่า (กลุ่ม)

2. กำหนดจุดศูนย์กลางเริ่มต้น K จุด เรียกว่า cluster centers หรือ Centroid

3. นำวัตถุทั้งหมดจัดเข้ากลุ่ม โดยทำการหาค่าระยะห่างระหว่างข้อมูล กับจุดศูนย์กลาง หากข้อมูลไหนใกล้ค่าจุดศูนย์กลางตัวไหนที่สุดอยู่ กลุ่มนั้น

4. หาค่าเฉลี่ย แต่ละกลุ่ม ให้เป็นค่าจุดศูนย์กลางใหม่ ( new centroid )

5. ทำซ้ำข้อ 2 จนกระทั่งค่าเฉลี่ยหรือจุดศูนย์กลางในแต่ละกลุ่มจะไม่ เปลี่ยนแปลงจึงหยุดทำ

```
# สร้าง function ในการรวมข้อมูลทั้ง 3 ชุดเข้าด้วยกัน

def preprocess_purchases_prices_distances():

  # join df_purchases และ df_prices  โดยใช้ product_id เป็น key
  output = pd.merge(df_purchases, df_prices, how='left', on='product_id')
  output['amount'] = output['quantity']*output['price']

  # join df_purchases, df_prices และ df_distances โดยใช้ customer_id, shop_id เป็น key
  output = pd.merge(output, df_distances, how='left', on=['customer_id', 'shop_id'])

  # สร้าง column ใหม่เพื่อให้รู้ว่าข้อมูล 1 row คือ 1 transaction หรือ 1 ธุรกรรม
  # txn มักใช้เป็นคำย่อของคำว่า transaction
  output['cnt_txn'] = 1

  return output
```

```
# สร้าง function ในการเตรียมข้อมูลให้อยู่ในรุปของ customer profile

def preprocess_trans_df(df, value_col, func='sum', total_only=False, shop_id_only=False):


  # เงื่อนไข parameter ต่างๆ ใน pivot_table
  if func == 'sum':
    suffix = 'total'
  elif func == 'mean':
    suffix = 'avg'
  else:
    suffix = func

  # by shop id
  pv1 = pd.pivot_table(df, values=value_col, index='customer_id', \
                      columns='shop_id', aggfunc=func)

  pv1.fillna(0, inplace=True)
  pv1.columns = [f'shop_id{i}_{value_col}_{suffix}' for i in pv1.columns.values]
  pv1.reset_index(inplace=True)

  # total
  df_temp = df.copy()
  df_temp['shop_id'] = 'all'
  pv2 = pd.pivot_table(df_temp, values=value_col, index='customer_id', \
                    columns='shop_id', aggfunc=func)

  pv2.fillna(0, inplace=True)
  pv2.columns = [f'all_shops_{value_col}_{suffix}']
  pv2.reset_index(inplace=True)


  if shop_id_only:
    return pv1
  elif total_only:
    return pv2
  else:
    output = pd.merge(pv1, pv2, on='customer_id', how='left')
    return output
```

```
def extract_feat():


  df_trans1 = preprocess_trans_df(df_trans, value_col='amount', func='sum')
  df_trans2 = preprocess_trans_df(df_trans, value_col='amount', func='mean')

  df_trans3 = preprocess_trans_df(df_trans, value_col='quantity', func='sum')
  df_trans4 = preprocess_trans_df(df_trans, value_col='quantity', func='mean')

  df_trans5 = preprocess_trans_df(df_trans, value_col='price', func='mean')

  df_trans6 = preprocess_trans_df(df_trans, value_col='distance', func='mean', total_only=True)
  df_trans7 = preprocess_trans_df(df_trans, value_col='distance', func='min', total_only=True)
  df_trans8 = preprocess_trans_df(df_trans, value_col='distance', func='max', total_only=True)

  df_trans9 = preprocess_trans_df(df_distances, value_col='distance', func='mean', shop_id_only=True)

  df_trans10 = preprocess_trans_df(df_trans, value_col='cnt_txn', func='sum', shop_id_only=True)
  df_trans10['shop_used'] = preprocess_trans_df(df_trans, value_col='cnt_txn', func='max', shop_id_only=True).drop('customer_id', axis=1).sum(axis=1)

  output = pd.merge(df_trans1, df_trans2, how='left', on='customer_id')
  output = pd.merge(output, df_trans3, how='left', on='customer_id')
  output = pd.merge(output, df_trans4, how='left', on='customer_id')
  output = pd.merge(output, df_trans5, how='left', on='customer_id')
  output = pd.merge(output, df_trans6, how='left', on='customer_id')
  output = pd.merge(output, df_trans7, how='left', on='customer_id')
  output = pd.merge(output, df_trans8, how='left', on='customer_id')
  output = pd.merge(output, df_trans9, how='left', on='customer_id')
  output = pd.merge(output, df_trans10, how='left', on='customer_id')

  output.drop('customer_id', axis=1, inplace = True)

  return output[sorted(output.columns)]
```

# Model K-Mean Clustering

**Customer Segmentation ด้วย K-means Clustering**

ใช้ K-Means Clustering Algorithm ในการแบ่งกลุ่มลูกค้าข้อมูลตัวอย่าง
โดยที่เราจะต้องกำหนดค่า K ซึ่งเป็น Hyperparameter ให้กับ Algotithm
หรือ
อาจจะให้ Machine ช่วยเลือกค่า K ให้ว่าจะแบ่งลูกค้าเป็นกี่กลุ่มดี

เราจะใช้วิธีลองแบ่งกลุ่มไล่ตั้งแต่ 2-15 กลุ่ม แล้วหา
Within-cluster Sum of Squared Errors (WSS) หรือ Inertia ของ K นั้น ๆ มาพล็อตกราฟดูแนวโน้ม

Within-cluster Sum of Squared Errors (WSS) คือ ผลรวมของ ระยะระหว่างจุดข้อมูล กับ Centroid ที่อยู่ใน Cluster เดียวกัน ยกกำลังสอง

**Elbow methed**

```
from sklearn.cluster import KMeans
```

```
min_range = 2
max_range = 16

inertia = []
k_list = range(min_range, max_range)

for k in k_list:
    km = KMeans(n_clusters = k, random_state= 0)
    km.fit(feat_norm)
    score = km.inertia_
    inertia.append(score)


plt.figure(1 , figsize = (10 ,6))
# plt.plot(np.arange(min_range , max_range) , inertia , 'o')
plt.plot(np.arange(min_range , max_range) , inertia , '-' , alpha = 0.5)

plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

วิธีเลือกค่า K ก็คือ ให้สังเกตที่ข้อศอก Elbow ของกราฟหักศอกเยอะที่สุด
ในเคสนี้ จะเท่ากับ จะได้ K = 7 ถึง 8




**Auto Find K**

Source: https://jtemporal.com/kmeans-and-elbow-method/

```
def calculate_wcss(data):
    wcss = []
    for n in range(min_range, max_range):
        kmeans = KMeans(n_clusters=n,random_state=0)
        kmeans.fit(feat_norm)
        wcss.append(kmeans.inertia_)

    return wcss

from math import sqrt

def optimal_number_of_clusters(wcss):
    x1, y1 = min_range, wcss[0]
    x2, y2 = max_range, wcss[len(wcss)-1]

    distances = []
    for i in range(len(wcss)):
        x0 = i+2
        y0 = wcss[i]
        numerator = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominator = sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distances.append(numerator/denominator)

    return distances.index(max(distances)) + 2


# calculating the within clusters sum-of-squares for n cluster amounts
sum_of_squares = calculate_wcss(feat_norm)

# calculating the optimal number of clusters
n = optimal_number_of_clusters(sum_of_squares)
print('Number fo cluster =', n)
```

```
# Finding Elbow with Silhouette
from sklearn.metrics import silhouette_score

sil = []
kmax = 15

for k in range(4, kmax+1):
    print(k)
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(feat_norm)
    labels = kmeans.labels_
    sil.append(silhouette_score(feat_norm, labels, metric = 'euclidean'))
```

```
cluster = list(range(4, kmax+1))
plt.plot(cluster, sil)
plt.show()
```

## Radar chart

```
from math import pi

def plot_radar_df(df, feat_cols, label_col, selected_label_list, split_chart=False, dpi=100, n_cols = 3, fig_h=8, fig_w=8):

    # min-max normalization
    for col in feat_cols:
        df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

    # select data:
    row_idx = df[label_col].isin(selected_label_list)
    data = df[row_idx]
    data = data[feat_cols].values

    # number of variable
    N = len(feat_cols)

    # angle of each axis in the plot
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # set color
    my_color = plt.cm.get_cmap("Set2", len(df.index))

    # set font size and subplot
    if not split_chart:
        ax = plt.subplot(111, polar=True )
        size = 10
    else:
        size = 5
        dpi = 1.5*dpi

    for i in range(len(data)):

        title = label_col +'_' +str(selected_label_list[i])

        # Initialise the spider plot
        if split_chart:
            n_rows = np.ceil(len(df)/n_cols)
            ax = plt.subplot(n_rows,n_cols,i+1, polar=True )

        # If you want the first axis to be on top:
        ax.set_theta_offset(pi / 2)
        ax.set_theta_direction(-1)

        # Draw one axe per variable + add labels labels yet
        plt.xticks(angles[:-1], feat_cols, color='grey', size=int(size*0.8))

        # Draw ylabels
        ax.set_rlabel_position(0)
        plt.yticks([0.20, 0.4, 0.6, 0.8,  1], \
                   ["20%","40%","60%","80%","100%"], color="grey", size=size)
        plt.ylim(0,1)

        # Plot
        values = list(data[i,:])
        values += values[:1]
        ax.plot(angles, values, color=my_color(i), linewidth=2, linestyle='solid', label=title)
        ax.fill(angles, values, color=my_color(i), alpha=0.4)

        # Add a title or legend
        if split_chart:
            plt.title(title, size=11, color=my_color(i), y=1.2)
        else:
            plt.legend(loc='best', bbox_to_anchor=(0.1, 0.1))

        # Adjust DPI
        fig = plt.gcf()
        fig.set_dpi(dpi)
        fig.set_figheight(fig_h)
        fig.set_figwidth(fig_w)

    plt.tight_layout()
    plt.show()
```
